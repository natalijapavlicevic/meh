```python
import pygame
import random
import os
import sys

# Parameters
g = 9.81   # Acceleration due to gravity (m/s^2)
m = 5.24e-7  # Mass of the raindrop (kg)
kv = 1.85e-7  # Viscous drag coefficient (Nsm^-1)

class Rain(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.image = rain_img
        self.rect = self.image.get_rect()
        self.speedy = 0 #random.randint(5,25)
        self.rect.x = random.randint(-100,wn_width)
        self.rect.y = 0 #-wn_height

    def update(self):
        
        if self.rect.bottom > wn_height:
            self.speedy = 0 #random.randint(5,25)
            self.rect.x = random.randint(-wn_width,wn_width)
            self.rect.y = 0

        self.acceleration = g - (random.randint(97,103)*(kv/m)*self.speedy)/100
        self.speedy = self.speedy + self.acceleration
        self.rect.y = self.rect.y + self.speedy

# Initialize
pygame.init()

# Used to manage how fast the screen updates
clock = pygame.time.Clock()

# Define some colors
WHITE = (255, 255, 255)

# images
rain_img = pygame.image.load('rain.png')

# Set up window
wn_width = 700
wn_height = 1000
wn = pygame.display.set_mode((wn_width,wn_height))
pygame.display.set_caption('Torrential Rain')

# sprite block
rain_group = pygame.sprite.Group()

for i in range(1):
    rain = Rain()
    rain_group.add(rain)

running = True
# -------- Main Program Loop -----------
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False	
            pygame.quit()
            sys.exit()
            #quit()
    if running:
# Update
        rain_group.update()
# Draw all the spites
        wn.fill(WHITE)
        rain_group.draw(wn)
        pygame.display.flip()
        clock.tick(30)
        
pygame.quit()
quit()

# quit
```
