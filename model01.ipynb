```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from pylab import *

# BEZ OTPORA
g = 9.81   
m = 5.24e-7  #  masa kapljice
kv = 0  # NEMA OTPORA!

dt = 3

# pocetni uslovi
h = 1000 #m
v = 0
t = 20.0
n = int(round(t/dt))

H = []

for i in range(n-1):
    h += v * dt # staro v OJLEROVA METODA
    g += - (kv/m)*v # ovo ce uvek biti nula! tj g je const
    v -= g * dt # novo v

    H.append(h)


plt.figure(figsize=(5,5))
plt.scatter(np.ones_like(H),H)
plt.title('Bez otpora vazduha')
plt.ylabel('Visina [m]')
plt.ylim(0)
plt.show()
```
